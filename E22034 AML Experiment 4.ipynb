AIM : 
#### Use cars.csv with variables  MPG as Dependent variable and HP,WT,CYL,DISP as Independent Variable 
#### Draw 200 random samples of size 80
#### For each random sample 
            i. Fit a KNN Model
            ii.KNN Model with knn as 2
            iii.KNN Model with knn as 10
            iii.Make predictions for a fixed test input 
            iv. Study the variance of the output
            v. Conclude which one has a high variance model. 

```python
# Importing the dependencies 

import pandas as pd
import numpy as np
import random
from sklearn.neighbors import KNeighborsRegressor
```

```python
# mounting the drive 

from google.colab import drive
drive.mount('/content/drive')
```

```python
# reading the data from the google drive and reading as pandas dataset.

df = pd.read_csv('/content/drive/MyDrive/AML /cars.csv')
df.head() # Shows the first 5 rows of the dataset variable name df
```

```python
# Droping Variables which is not required for the experiment.
df = df.drop(["Car",'Acceleration','Model','Origin'], axis = True)
```

```python
# Seperating the independent and dependent variables 
X = df.drop(['MPG'], axis = 1 ) # All X's values 
y = df.MPG # y - values 
```

# Taking 200 random samples with 80 observations and fitting a KNN model with k nearest neibghours as 2 and making predictions for fixed test input.

```python
# Modeling the KNN Regressor 

predicted_knn_2 = [] # Initializing a empty list.


for i in range (1,201): # initiating a loop for 200 times.
    
    test = X.sample(1,random_state = 123) # taking a fixed sample.
    train_X = X.sample(80) # Taking 80 observations for train in X.
    train_y = y.sample(80) # Taking 80 observations for test data in y.
    
    # Step-1: instantiate learning model.
    knn = KNeighborsRegressor(n_neighbors=2, metric='euclidean')  # Fitting  a KNN Model with knn as 2

    # Step-2: fitting the model.
    knn.fit(train_X, train_y)

    # Step-3: Predicting the values.
    predicted_values_2 = knn.predict(test) # Making predictions for a fixed test input.
    predicted_knn_2.append(predicted_values_2) # Appending the predictions into predicted_knn_2 list.

np.var(predicted_knn_2) # Calculating  the variance of the output for the KNN Model.
```

# Taking 200 random samples with 80 observations and fitting a KNN model with k nearest neibghours as 10 and making predictions for fixed test input.

```python
# Modeling the KNN Regressor 

predicted_knn_10 = [] # Initializing a empty list.


for i in range (1,201): # initiating a loop for 200 times.
    
    test = X.sample(1,random_state = 123) # taking a fixed sample.
    train_X = X.sample(80) # Taking 80 observations for train in X.
    train_y = y.sample(80) # Taking 80 observations for test data in y.
    
    # Step-1: instantiate learning model.
    knn = KNeighborsRegressor(n_neighbors=10, metric='euclidean')  # Fitting  a KNN Model with knn as 10

    # Step-2: fitting the model.
    knn.fit(train_X, train_y)

    # Step-3: Predicting the values.
    predicted_values_10 = knn.predict(test) # Making predictions for a fixed test input.
    predicted_knn_10.append(predicted_values_10) # Appending the predictions into predicted_knn_10 list.

np.var(predicted_knn_10) # Calculating  the variance of the output for the KNN Model.
```
