```python
# Droping Variables which is not required for the experiment just keeping MPG and Weight
df = df.drop(["Car",'Acceleration','Model','Origin','Cylinders','Displacement','Horsepower'], axis = True)
```

```python
# Seperating the independent and dependent variables 
X = df.Weight # is taken as X
y = df.MPG # is taken as y
X.shape
```

## Fitting a Linear regression model to predict MPG based on the weight of the cars 


```python
# To predict the estimated values :

X_Sum= np.sum(X)
y_Sum = np.sum(y)
X_sqr_Sum = np.sum(X**2)
X_Sum_2 = X_Sum**2
X_y_Sum = np.sum(X * y)
n = len(X)
n
```

```python
# finding the value of bo and b1 using the formula 

bo = ((y_Sum * X_sqr_Sum) - (X_Sum * X_y_Sum)) / ((n * X_sqr_Sum) - (X_Sum_2)) # Intercept of y 
b1 = ((n * X_y_Sum) - (X_Sum * y_Sum)) / ((n * X_sqr_Sum) - (X_Sum_2)) # Slope of y 
print(bo)
print(b1)
```

```python
# Fitting the Model 

Model_Predicted  = bo + b1 * X
Model_Predicted
```

```python
# Error in  the model 

Error = y - Model_Predicted

# MAE (Mean Absolute Error)
np.mean(np.abs(Error)) # The Average of the absolute deviation between the Actual value and the predicted value of each observations.
```

# Conclusion
